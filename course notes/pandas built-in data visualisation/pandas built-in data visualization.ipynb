```python
# pandas built-in visualisation was built on top matplotlib and allows 
# you to call data visualisation directly off a data frame
```

```python
import pandas as pd
import numpy as np
import seaborn as sns # to make the plots more presentable with beautiful styles
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
# read csv data
df1 = pd.read_csv('df1', index_col=0) # index is the first column
df1.head()
```

```python
# df1.rename(columns={ df1.columns[0]: "index" })

df1.rename_axis('index')
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

```python
# to directly plot a histogram of df1's column A
df1.A.hist(figsize=(12, 5))
```

```python
# OR

df1['A'].hist(figsize=(12, 5))
```

```python
# pandas calls matplotlib under the hood
# you can then use matplotlib arguments to futher enhances the plots

df1.A.hist(bins=30, figsize=(12, 5))
```

```python
# other panda plots can be done by calling plot on a column of 
# the dataframe and then specifying the plot kind argument

df1.A.plot(kind='kde', figsize=(12, 5))
```

```python
df1.A.plot(kind='hist', bins=30, figsize=(12, 5))
```

```python
 # OR
    
df1.A.plot.hist(bins=30, figsize=(12, 5))
```

```python
df2.head()
```

```python
# area plot
df2.plot.area(figsize=(12, 5))
```

```python
df2.plot.area(alpha=0.5, figsize=(12, 5))
```

```python
# bar plot

df2.plot.bar(figsize=(12, 5))
```

```python
df2.plot.bar(x='b', y='a', figsize=(12, 5))
```

```python
# the index of df2 is categorical because a group of values (not a single value) occur at same index
# therefore, the values in each categgory can be stacked

df2.plot.bar(stacked=True, figsize=(12, 5))
```

```python

df2.plot.hist(bins=40, figsize=(12, 5))
```

```python
# line plots: x and y values must be specified

df1.plot.line(use_index=True, y='B', figsize=(12, 5), lw=2) # here the data frame index is set as the x values, lw = line width

#dataframe.index is no longer accessible normally but 
#use_index=True makes it possible to use the index values in plots
```

```python
df2.plot.line(x='d', y='c')
```

```python
df1.plot.scatter(x='D', y='B', c='A', figsize=(12, 5), cmap='viridis') # c = color
```

```python
# size argument can be specified for the scatter. there won't be need to specify c
df1.plot.scatter(x='A', y='B', figsize=(12, 5), s=df1['C']*50)
```

```python
df2.plot.box(figsize=(12, 5), cmap='viridis')
```

```python
# hexagonal bin plot
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.head()
```

```python
df.plot.hexbin(x='a', y='b', gridsize=25, figsize=(12, 8), cmap='coolwarm') # gridsize modifies the hexbin plots
```

```python
# column density estimation

df2.a.plot.density(figsize=(12, 8)) # df2.a.plot.kde(figsize=(12, 8))
```

```python
# density of all the variables (columns)
df2.plot.density(figsize=(12, 8))
```
