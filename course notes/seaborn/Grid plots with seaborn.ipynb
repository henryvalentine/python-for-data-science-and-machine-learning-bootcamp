```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set()
colpalette = sns.color_palette()
sns.palplot(colpalette)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
# grid plots can be produced by calling PairGrid
# it will take the columns and present 3 level-grids of their pairs: map_upper, map_diagonal, and map_lower 
# when a plot type is piped to the PairGrid object, it plots that same type in all the grid levels
# different plot types can be specified for the 3 different levels to make more meaningful presentation

gridPlot = sns.PairGrid(iris, hue="species")
gridPlot.map_diag(sns.distplot)
gridPlot.map_upper(plt.scatter)
gridPlot.map_lower(sns.kdeplot)
gridPlot.add_legend();
```

```python
# FacetGrid is used to visualise the distribution of a variable or the relationships between 
# multiple variables within subsets of a dataset
# It can be plotted with upto three dimensions: row, col, and hue
# The row and col variables result in array of axes while the hue variable is a 
# third dimension along the dept axis where different levels are plotted with different colors

tips = sns.load_dataset('tips')
tips.head()
```

```python
f = sns.FacetGrid(tips, 'time', 'smoker', hue='tip', aspect=0.8, height=8)
f.map(plt.scatter, 'total_bill', 'tip')
```
