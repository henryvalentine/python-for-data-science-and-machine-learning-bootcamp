```python
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(5)
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.jointplot(x='total_bill', y='tip',data=tips) #scatter plot by default
```

```python
sns.jointplot(x='total_bill', y='tip',data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip',data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip',data=tips, kind='kde')
```

```python
sns.pairplot(tips, 'sex') # A nice way to quickly visualise the data
```

```python
sns.rugplot(tips['total_bill']) # Draws a single line for every data entry in the dataset column being considered
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.barplot(x='tip', y='total_bill', data=tips)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips,hue='smoker')
```

```python
# A violin plot is the same with barplot but contains 
# a little more information more than a boxplot

sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
# To plot the information (eg. to place the dataset column for which the hue attribute is set)
# side-by-side within one violin, use the split option as shown below

sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split='sex')
```

```python
# A stripplot is a form of scatter that stacks the points upon each othe

sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
# adding jitter=True adds some sort of noise and thus helps to separate the stacked points
# so they can be clearer

sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', jitter=True, dodge=True) # split parameter was renamed to dodge
```

```python
# A swarm plot is similar to stripplot but it presents the plot points such that they are not stacked together
# thereby making it easy for the plot to be interpreted clearly
# it is an essential combination of strip and violin plots combined together

sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)

# the points are clearly represented.
# the downside to swarmplot is that it doesn't sometimes scale to very large numbers
```

```python
# violinplot can be combined with swarmplot to clearly show the pieces of information

sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', color='cyan')
```

```python
# factor plot has been renamed to catplot
# it is the most generally used plot type

sns.catplot(x='day', y='total_bill', data=tips, kind='bar',hue='sex')
```
