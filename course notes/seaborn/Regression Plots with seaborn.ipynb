```python
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
sb.set()
```

```python
tips = sb.load_dataset('tips')
tips.head()
```

```python
# the goal of seaborn is to make exploration through the visualisation of a dataset
# quick and easy
# regressional plots in seaborn are primarily intended to emphasise patterns within a dataset
# during exploratory data analyses

sb.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], scatter_kws={'s': 100})

# adding scatter_kws={'s': 100} from matplotlib modifies the sizes of the scatter plots
```

```python
# gridplots can be produced ontop regression plots by using 
# col='a variable of the dataset' and row='a different variable of the dataset'
# lmplot = linear model plot

sb.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

```python
# aspect ratios, hue and height (formerly size) can be added to the plots to make their presentations clearer 

sb.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time', hue='sex', aspect=0.9, height=8)
```
