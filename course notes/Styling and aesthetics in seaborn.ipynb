```python
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline

# plot background can be set by default by calling sb.set()
# the parameters include whitegrid, darkgrid, ticks, dark, white 

sb.set_style('whitegrid')
```

```python
tips = sb.load_dataset('tips')
tips.head()
```

```python
sb.countplot(x='sex', data=tips)
```

```python
# matplotlib can be used to set the sizes of plot figures
plt.figure(figsize=(14, 8)) # figsize is a tupple
sb.countplot(x='sex', data=tips)
```

```python
# seaborn comes with set_context that helps to override default parameters and 
# change font scales depending on the need
# the set_context has parameters which include: paper, notebook, talk, poster
# if the context is set to poster for instance, the produced plot figure will
# be larger and the fonts also scaled up. seaborn assumes the plot will be published on a poster

sb.set_context('poster')
sb.countplot(x='sex', data=tips)
```

```python
# the font size can be scaled up further by adding a second argument font_scale in the plot

sb.set_context('poster', font_scale=2)
sb.countplot(x='sex', data=tips)
```

```python
# comprehensive list of color maps/paletts can be accessed by performing 'matplotlib colormap' search on google
# this will bring up links among which will be the matplotlib color palette documentation page

sb.countplot(x='sex', data=tips, palette='seismic')
```
