# Pandas Data Visualization Exercise

This is just a quick exercise for you to review the various plots we showed earlier. Use **df3** to replicate the following plots. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
sns.set_style=('ticks')
```

```python
df3.info()
```

```python
df3.head()
```

** Recreate this scatter plot of b vs a. Note the color and size of the points. Also note the figure size. See if you can figure out how to stretch it in a similar fashion. Remeber back to your matplotlib lecture...**

```python
df3.plot.scatter(x='a', y='b', figsize=(12,3), color='r', s=50, edgecolor='black', lw=1) # s = size of the dot
```

** Create a histogram of the 'a' column.**

```python
sns.set_style = ('ticks')
df3.a.hist(color='b', grid=False)
```

** These plots are okay, but they don't look very polished. Use style sheets to set the style to 'ggplot' and redo the histogram from above. Also figure out how to add more bins to it.***

```python
plt.style.use('ggplot') # set the style to ggplot
df3.a.hist(color='red', bins=30, alpha=0.34)
# sns.set_style('darkgrid')
```

** Create a boxplot comparing the a and b columns.**

```python
#sns.set_style=('darkgrid')
plt.style.use = ('ggplot') # set the style to ggplot
df3[['a','b']].plot.box(color=dict(boxes='r', whiskers='red',medians='darkblue', caps='none'), 
                            whiskerprops = dict(linestyle='--', linewidth=1.0),   
                            boxprops = dict(linestyle='-', linewidth=1))
```

** Create a kde plot of the 'd' column **

```python
df3.d.plot.density(c='r', alpha=0.5)
```

** Figure out how to increase the linewidth and make the linestyle dashed. (Note: You would usually not dash a kde plot line)**

```python
df3.d.plot.density(c='r', lw=4, ls='--', alpha=0.6)
```

** Create an area plot of all the columns for just the rows up to 30. (hint: use .ix).**

```python
df3.head()
```

```python
# the .ix function is used for indexing/slicing of collections like data frame
# .ix is deprecated and replaced by .loc for label based indexing OR .iloc for positional based indexing

d = df3.iloc[0:30] # This is the same thing as: d = df3[0:30]
d.plot.area(alpha=0.4)
```
